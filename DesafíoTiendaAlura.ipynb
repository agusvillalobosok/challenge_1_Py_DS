### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
precio_tienda_1 = tienda["Precio"].tolist()

precio_tienda_2 = tienda2["Precio"].tolist()

precio_tienda_3 = tienda3["Precio"].tolist()

precio_tienda_4 = tienda4["Precio"].tolist()


ingreso_total_tienda_1 = sum(precio_tienda_1)

ingreso_total_tienda_2 = sum(precio_tienda_2)

ingreso_total_tienda_3 = sum(precio_tienda_3)

ingreso_total_tienda_4 = sum(precio_tienda_4)
```

```python
### Gráfico Índice de ingresos por tienda.
import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette("pastel")

# Datos de ejemplo (en millones)
tiendas = ["Tienda Alura", "Tienda Alura 2", "Tienda Alura 3", "Tienda Alura 4"]
ingresos = [ingreso_total_tienda_1, ingreso_total_tienda_2, ingreso_total_tienda_3, ingreso_total_tienda_4]

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas, ingresos, color=colors, width=0.1)

# Detalles del gráfico
plt.title("Índice de ingresos por tienda", fontsize=12, fontweight='bold')
plt.xlabel("")
plt.ylabel("Ingresos (en millones)", fontsize=10, fontweight='bold')
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.yticks([])  # Oculta los números del eje Y

# Opcional: quitar la línea del eje Y (borde izquierdo)
plt.gca().spines['left'].set_visible(False)

# Mostrar valores encima de las barras en formato corto (1 decimal y "M")
for bar in bars:
    height = bar.get_height()
    abreviado = height / 1000000  # Convertir a miles de millones si están en millones
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.05 * max(ingresos),
        f'{abreviado:.1f}M',
        ha='center', va='bottom', fontsize=9
    )

plt.show()
```

# 2. Ventas por categoría

```python
### Gráfico de Productos vendidos por las tiendas Alura.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos
porcentajes = [vendidos_x_categoria_T1, vendidos_x_categoria_T2, vendidos_x_categoria_T3, vendidos_x_categoria_T4]

# Colores suaves de seaborn
colors = sns.color_palette('pastel')



# Datos de ejemplo
categorias = vendidos_x_categoria_T1.index  #  categories
ventas = vendidos_x_categoria_T1.values #  counts
ventas_2 = vendidos_x_categoria_T2.values
ventas_3 = vendidos_x_categoria_T3.values
ventas_4 = vendidos_x_categoria_T4.values
# Crear gráfico
plt.figure(figsize=(16, 5))
# Use the first color from the palette
plt.plot(categorias, ventas, marker='o',label="Tienda Alura", color=colors[0], linewidth=2)
plt.plot(categorias, ventas_2, marker='s',label="Tienda Alura 2", color=colors[1], linewidth=2)
plt.plot(categorias, ventas_3, marker='.',label="Tienda Alura 3", color=colors[2], linewidth=2)
plt.plot(categorias, ventas_4, marker='^',label="Tienda Alura 4", color=colors[3], linewidth=2)


# Etiquetas
plt.title("Productos vendidos por las tiendas Alura", fontsize=14, fontweight='bold')
plt.xlabel("Categorías de los articulos")
plt.ylabel("Cantidad de ventas")
plt.grid(True)
plt.legend()  # Muestra la leyenda

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda_1 = round(tienda["Calificación"].mean(),2)
print(f"Esta es la calificación promedio que dieron los clientes de la tienda Alura: {calificacion_tienda_1}")

calificacion_tienda_2 = round(tienda2["Calificación"].mean(),2)
print(f"Esta es la calificación promedio que dieron los clientes de la tienda Alura 2: {calificacion_tienda_2}")

calificacion_tienda_3 = round(tienda3["Calificación"].mean(),2)
print(f"Esta es la calificación promedio que dieron los clientes de la tienda Alura 3: {calificacion_tienda_3}")

calificacion_tienda_4 = round(tienda4["Calificación"].mean(),2)
print(f"Esta es la calificación promedio que dieron los clientes de la tienda Alura 4: {calificacion_tienda_4}")
```

```python
### Gráfico de calificación tienda Alura.
import matplotlib.pyplot as plt

# Datos de ejemplo
tiendas = ["Tienda Alura", "Tienda Alura 2", "Tienda Alura 3", "Tienda Alura 4"]
calificaciones = [calificacion_tienda_1, calificacion_tienda_2, calificacion_tienda_3, calificacion_tienda_4]

# Crear gráfico horizontal
plt.figure(figsize=(10, 5))
bars = plt.barh(tiendas, calificaciones, color='green')

# Etiquetas
plt.title("Calificación de Tienda Alura", fontsize=14, fontweight='bold')
plt.xlabel("Satisfacción del cliente", fontsize=12)
plt.ylabel("")

# Agregar valores centrados sobre cada barra
for bar in bars:
    width = bar.get_width()
    plt.text(
        width - 0.1,  # Un poco antes del final de la barra
        bar.get_y() + bar.get_height() / 2,
        f"{width:.2f}",  # Un decimal
        va='center', ha='right', color='white', fontsize=10, fontweight='bold'
    )

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python


ventas_productos_T1 = tienda["Producto"].value_counts()
max_vendido_T1 = ventas_productos_T1.idxmax()
min_vendido_T1 = ventas_productos_T1.idxmin()


ventas_productos_T2 = tienda2["Producto"].value_counts()
max_vendido_T2 = ventas_productos_T2.idxmax()
min_vendido_T2 = ventas_productos_T2.idxmin()


ventas_productos_T3 = tienda3["Producto"].value_counts()
max_vendido_T3 = ventas_productos_T3.idxmax()
min_vendido_T3 = ventas_productos_T3.idxmin()


ventas_productos_T4 = tienda4["Producto"].value_counts()
max_vendido_T4 = ventas_productos_T4.idxmax()
min_vendido_T4 = ventas_productos_T4.idxmin()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
# Colores suaves de seaborn
colors = sns.color_palette('pastel')

# Primero sacamos las cantidades vendidas de esos productos
cantidad_max_T1 = ventas_productos_T1[max_vendido_T1]
cantidad_min_T1 = ventas_productos_T1[min_vendido_T1]

cantidad_max_T2 = ventas_productos_T2[max_vendido_T2]
cantidad_min_T2 = ventas_productos_T2[min_vendido_T2]

cantidad_max_T3 = ventas_productos_T3[max_vendido_T3]
cantidad_min_T3 = ventas_productos_T3[min_vendido_T3]

cantidad_max_T4 = ventas_productos_T4[max_vendido_T4]
cantidad_min_T4 = ventas_productos_T4[min_vendido_T4]

# Ahora organizamos los datos
productos = [max_vendido_T1, min_vendido_T1, max_vendido_T2, min_vendido_T2,
             max_vendido_T3, min_vendido_T3, max_vendido_T4, min_vendido_T4]

cantidades = [cantidad_max_T1, cantidad_min_T1, cantidad_max_T2, cantidad_min_T2,
              cantidad_max_T3, cantidad_min_T3, cantidad_max_T4, cantidad_min_T4]

colores = ['green', 'red', 'skyblue', 'red', 'blue', 'red', 'purple', 'red']

# Creamos el gráfico
plt.figure(figsize=(12, 6))
plt.bar(productos, cantidades,label="Tienda Alura", color=colors[0])
plt.bar(productos, cantidades,label="Tienda Alura 2", color=colors[2])
plt.bar(productos, cantidades,label="Tienda Alura 3", color=colors[4])
plt.bar(productos, cantidades,label="Tienda Alura 4", color=colors[6])
plt.bar(productos, cantidades, color=colors)

# Agregamos detalles
plt.title("Productos más vendidos y menos vendidos por tienda", fontweight='bold' )
plt.xlabel("Productos")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Mostrar gráfico
plt.tight_layout()
plt.show()

```

# 5. Envío promedio por tienda

```python
costo_envio_T1 = round(tienda["Costo de envío"].mean(),2)
print(f"El costo de envío en promedio de la tienda Alura fue {costo_envio_T1}." )

costo_envio_T2 = round(tienda2["Costo de envío"].mean(),2)
print(f"El costo de envío en prodedio de la tienda Alura 2 fue {costo_envio_T2}")

costo_envio_T3 = round(tienda3["Costo de envío"].mean(),2)
print(f"El costo de envío en prodedio de la tienda Alura 3 fue {costo_envio_T3}")

costo_envio_T4 = round(tienda4["Costo de envío"].mean(),2)
print(f"El costo de envío en prodedio de la tienda Alura 4 fue {costo_envio_T4}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo general
sns.set(style="whitegrid")  # Fondo claro y limpio

# Datos de ejemplo
costos_envio = [costo_envio_T1, costo_envio_T2, costo_envio_T3, costo_envio_T4]
ingresos = [ingreso_total_tienda_1, ingreso_total_tienda_2, ingreso_total_tienda_3, ingreso_total_tienda_4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]  # Opcional: para etiquetas

# Crear figura
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    costos_envio, ingresos,
    color=sns.color_palette("deep")[0],
    s=200, alpha=0.8,
    edgecolors='black',
    linewidth=1
)

# Título y etiquetas
plt.title("Relación entre los ingresos y costos de envío\nTiendas Alura", fontsize=16, fontweight='bold')
plt.xlabel("Costo promedio de envío ($)", fontsize=12)
plt.ylabel("Ingresos (en millones)", fontsize=12)

# Anotar cada punto con el nombre de la tienda
for i, tienda in enumerate(tiendas):
    plt.text(
        costos_envio[i] + 0.05, ingresos[i],  # Ajuste de posición
        tienda,
        fontsize=10,
        ha='left',
        va='center'
    )

# Opcional: eliminar bordes innecesarios
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```
